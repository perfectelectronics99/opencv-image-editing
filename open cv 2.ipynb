### EDGE  DETECTIONS

```python
import numpy as np 
import cv2 
from matplotlib import pyplot as plt 


img = cv2.imread('education.jpg') 

# convert image to gray scale image 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 

# detect corners with the goodFeaturesToTrack function. 
corners = cv2.goodFeaturesToTrack(gray, 35, 0.01, 20) 
corners = np.int0(corners) 

# we iterate through each corner, 
# making a circle at each point that we think is a corner. 
for i in corners: 
	x, y = i.ravel() 
	cv2.circle(img, (x, y), 3, 255, -1) 

plt.imshow(img), plt.show() 
```

### cv2.INTER_AREA is used for shrinking, whereas cv2.INTER_CUBIC  # is used for zooming. 

```python
import cv2 
import numpy as np 

FILE_NAME = 'education.jpg'
try: 
	# Read image from disk. 
	img = cv2.imread(FILE_NAME) 

	# Get number of pixel horizontally and vertically. 
	(height, width) = img.shape[:2] 

	res = cv2.resize(img, (int(width / 1), int(height / 1)), interpolation = cv2.INTER_CUBIC) 

	# Write image back to disk. 
	cv2.imwrite('result.jpg', res) 

except IOError: 
	print ('ERROR !!!') 
```

```python
import cv2 
import numpy as np 

NAME = 'education.jpg'
try: 
	# Read image from the disk. 
	img = cv2.imread(NAME) 

	# Shape of image in terms of pixels. 
	(rows, cols) = img.shape[:2] 

	# getRotationMatrix2D creates a matrix needed for transformation. 
	# We want matrix for rotation w.r.t center to 45 degree without scaling. 
	M = cv2.getRotationMatrix2D((cols /2, rows / 2), 360, 1) 
	res = cv2.warpAffine(img, M, (cols, rows)) 

	# Write image back to disk. 
	cv2.imwrite('OUT.jpg', res)
    
except IOError: 
	print ('Error while reading files !!!')
```
